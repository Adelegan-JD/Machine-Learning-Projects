## Data Cleaning

```python
# Check for missing values
df.isna().sum()
```

```python
# Check for duplicates
duplicates = df.duplicated().sum()

if duplicates.is_integer():
    print(f'There are {duplicates} duplicates in this dataset')
else:
    print('There are no duplicates in this dataset')
```

```python
# Remove Dupliactes
df = df.drop_duplicates()

print(f'{duplicates} duplicates have been dropped')
```
